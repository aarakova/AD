Первым делом читаем файл с помощью команды read.csv и выводим таблицу.

````python
import pandas as pd

df = pd.read_csv('credit_risk_lab1_v15.csv')
df.head(20)
````

Далее необходимо удостовериться в корректности названия столбцов и их соответствия данным варианта. Проверить датасет на наличие дубликатов с помощью метода ```duplicated()```, посчитать и вывести количество дубликатов, а также избавиться от них с помощью метода ```drop_duplicates()```.

````python
duplicated_rows = df.duplicated()
print(duplicated_rows.sum())
````

````python
duplicated_rows = df.duplicated()
df = df.drop_duplicates()
print(duplicated_rows.sum())
````

После удаления строк-дубликатов обновили индексацию используя метод ```reset_index()```, который создает новый датафрейм, в котором индексы исходного датафрейма станут новой колонкой с названием index, а все строки получают обычные индексы без пропусков. В этом методе использовали аргумент ```drop```, чтобы не создавать столбец ```index```.

````python
df = df.drop_duplicates().reset_index(drop=True)
df.head(20)
````

Следующим шагом с помощью метода ```info()``` получаем общую информацию о столбцах, а также проверяем датафрейм на наличие неявных дубликатов с помощью метода ```unique()```, который возвращает перечень уникальных значений в столбце, и метода ```value_counts()```, который анализирует столбец и показывает каждое уникальное значение, прия этом подсчитывая частоту его встречаемости в списке.

````python
df.info()
````

````python
print(df['Home'].unique())
print(df['Intent'].unique())
print(df['Default'].unique())
````

````python
df.describe()
````

В столбцах:

•	'Age' что есть два значения равных 144 и 123, которые необходимо заменить с помощью метода replace() на NaN и 23;

•	'Home' есть значение RENET, которое необходимо заменить на RENT;

•	'Emp_length' есть значения 123 (будет заменено на 23) и nan (будет заменено на 0);

•	'Rate' нет нулевых и отрицательных значений, но есть значения некорректного формата. Необходимо привести все значения к одному формату (float64);

•	'Percent_income' необходимо привести все значения к формату float64;

•	'Default' необходимо заменить значение No на N.

````python
df['Age'] = df['Age'].replace(144, pd.NA)
df['Age'] = df['Age'].replace(123, 23)

df['Home'] = df['Home'].replace('RENET', 'RENT')

df['Emp_length'] = df['Emp_length'].replace(123, 23)
df['Emp_length'] = df['Emp_length'].replace('nan', 0)

df['Default'] = df['Default'].replace('No', 'N')
````

Столбцы 'Rate' и 'Percent_income' приводим к float64 с помощью метода ```to_numeric()``` с аргументом ```errors='cource'``` (некорректные значения принудительно заменяются на NaN).

````python
df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')
df['Percent_income'] = pd.to_numeric(df['Percent_income'], errors='coerce')
````

Далее необходимо посчитать пропуски с помощью метода ```isna()```

````python
df.isna().sum()
````

Удаляем строки с пропущенными значениями с помощью метода ```dropna()```

````python
df = df.dropna()
df.isna().sum()
````

**Задание 3**

Сводная таблица (pivot_table) - средняя сумма кредита amount для каждой цели кредита. Отсортировать по убыванию. Округлить до трёх знаков.
